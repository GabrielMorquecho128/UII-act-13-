## 43 Pago automático recurrente

```python
# ======================================================
# 🏦 CBTis 128
# 📚 Elabora Proyectos de Programacion Logica
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Eliseo Nava
# 👨‍🎓 Alumno: Marco Morquecho
# 🏦 Proyecto: Pago Automatico Recurrente IA
# ======================================================

class Pago_Automatico_Recurrente:

    def __init__(self):

        self.pagos = []

    # ======================================================
    # ➕ AGREGAR PAGO
    # ======================================================

    def agregar_pago(self):

        nombre = input("Ingrese nombre del cliente: ")

        cuenta = input("Ingrese numero de cuenta: ")

        saldo = int(input("Ingrese saldo disponible: "))

        monto = int(input("Ingrese monto del pago: "))

        servicio = input("Ingrese servicio: ")

        datos = {

            "Nombre": nombre,
            "Cuenta": cuenta,
            "Saldo": saldo,
            "Monto": monto,
            "Servicio": servicio

        }

        self.pagos.append(datos)

        print("\nPago agregado correctamente")

    # ======================================================
    # 📋 LISTAR PAGOS
    # ======================================================

    def listar_pagos(self):

        print("\n==============================================================")
        print("               LISTA DE PAGOS AUTOMATICOS")
        print("==============================================================")

        if len(self.pagos) == 0:

            print("No hay pagos registrados")

        else:

            for datos in self.pagos:

                print("--------------------------------------------------------------")

                print("Nombre    :", datos["Nombre"])
                print("Cuenta    :", datos["Cuenta"])
                print("Saldo     :", datos["Saldo"])
                print("Monto     :", datos["Monto"])
                print("Servicio  :", datos["Servicio"])

    # ======================================================
    # ✏️ ACTUALIZAR PAGO
    # ======================================================

    def actualizar_pago(self):

        nombre = input("Ingrese nombre del cliente: ")

        encontrado = False

        for datos in self.pagos:

            if datos["Nombre"] == nombre:

                datos["Monto"] = int(input("Nuevo monto: "))

                datos["Servicio"] = input("Nuevo servicio: ")

                print("\nPago actualizado correctamente")

                encontrado = True

        if encontrado == False:

            print("Cliente no encontrado")

    # ======================================================
    # ❌ BORRAR PAGO
    # ======================================================

    def borrar_pago(self):

        nombre = input("Ingrese nombre del cliente: ")

        encontrado = False

        for datos in self.pagos:

            if datos["Nombre"] == nombre:

                self.pagos.remove(datos)

                print("\nPago eliminado correctamente")

                encontrado = True

        if encontrado == False:

            print("Cliente no encontrado")

    # ======================================================
    # 🧠 ANALIZAR PAGO
    # ======================================================

    def analizar_pago(self):

        saldo = int(input("Ingrese saldo disponible: "))

        monto = int(input("Ingrese monto del pago: "))

        print("\nANALIZANDO PAGO...\n")

        if saldo >= monto:

            print("Pago aprobado correctamente")

        elif monto >= 5000:

            print("Pago alto detectado")

        else:

            print("Saldo insuficiente")

# ======================================================
# 📌 MENU PRINCIPAL
# ======================================================

sistema = Pago_Automatico_Recurrente()

while True:

    print("\n=========== MENU ===========")
    print("1. Agregar pago")
    print("2. Listar pagos")
    print("3. Actualizar pago")
    print("4. Borrar pago")
    print("5. Analizar pago")
    print("6. Salir")

    opcion = int(input("Seleccione una opcion: "))

    # ======================================================

    if opcion == 1:

        sistema.agregar_pago()

    # ======================================================

    elif opcion == 2:

        sistema.listar_pagos()

    # ======================================================

    elif opcion == 3:

        sistema.actualizar_pago()

    # ======================================================

    elif opcion == 4:

        sistema.borrar_pago()

    # ======================================================

    elif opcion == 5:

        sistema.analizar_pago()

    # ======================================================

    elif opcion == 6:

        print("\n===================================")
        print("      PROGRAMA FINALIZADO")
        print("===================================")

        print("Alumno: Marco Morquecho")
        print("Especialidad: Inteligencia Artificial")
        print("Materia: Elabora Proyectos de Programacion Logica")
        print("Docente: Eliseo Nava")
        print("CBTis 128")

        print("\nGracias por utilizar PAGO AUTOMATICO IA")

        break

    # ======================================================

    else:

        print("Opcion no valida")
```
